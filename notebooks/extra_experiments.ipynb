```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Crear el DataFrame con todos los datos
data = {
    'strategy': ['hyde', 'q2d', 'rag_fusion', 'fire_doc', 'fire_query', 'fire_answer', 'doc+answer'] * 3,
    'dataset': ['nfcorpus'] * 7 + ['fiqa'] * 7 + ['scifact'] * 7,
    'input': [27.33, 1713.82, 106.33, 2166.86, 2127.86, 2166.86, 4333.72,
              35.67, 1197.96, 114.67, 1294.59, 1261.59, 1294.59, 2589.18,
              47.64, 1922.96, 123.64, 2244.19, 2212.19, 2244.19, 4488.38],
    'output': [1713.82, 444.05, 169.42, 357.7, 57.41, 177.03, 534.73,
               2112.8, 461.42, 81.91, 381.33, 74.15, 140.91, 522.24,
               2528.56, 304.49, 107.96, 281.76, 61.48, 363.81, 645.57],
    'nDCG10': [38.1, 30.8, 34.8, 39.9, 32.8, 40.1, 40.6,
               45.9, 43.6, 39.1, 45.9, 37.0, 44.9, 45.9,
               75.5, 78.1, 72.6, 75.9, 68.9, 76.3, 75.6]
}
df = pd.DataFrame(data)

# 2. Calcular el costo por 1000 ejecuciones
cost_input_per_mil = 0.5
cost_output_per_mil = 1.5
df['cost_per_1k'] = ((df['input'] * cost_input_per_mil) + (df['output'] * cost_output_per_mil)) / 1000

# 3. Definir colores y formas
strategies = df['strategy'].unique()
datasets = df['dataset'].unique()

# Paleta de colores para las estrategias
color_palette = sns.color_palette('tab10', n_colors=len(strategies))
color_map = dict(zip(strategies, color_palette))

# Formas para los datasets
marker_map = {'nfcorpus': 'o', 'fiqa': 's', 'scifact': '^'}

# 4. Generar una gráfica para cada dataset
for dataset_name in datasets:
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(10, 7))

    # Filtrar el dataframe para el dataset actual
    df_subset = df[df['dataset'] == dataset_name]

    # Crear el gráfico de dispersión
    sns.scatterplot(
        data=df_subset,
        x='cost_per_1k',
        y='nDCG10',
        hue='strategy',
        palette=color_map,
        style='dataset',
        markers=marker_map,
        s=150, # Aumentar tamaño del marcador
        ax=ax
    )

    # Añadir etiquetas a cada punto para mayor claridad
    for i, row in df_subset.iterrows():
        ax.text(row['cost_per_1k'] + 0.03, row['nDCG10'], row['strategy'], fontsize=9)

    # Configuración de la gráfica
    ax.set_title(f'Costo vs. nDCG@10 para el Dataset: {dataset_name.upper()}', fontsize=16, weight='bold')
    ax.set_xlabel('Costo por 1,000 Ejecuciones ($)', fontsize=12)
    ax.set_ylabel('nDCG@10', fontsize=12)
    
    # Mover la leyenda fuera del área del gráfico
    ax.legend(title='Estrategias', bbox_to_anchor=(1.02, 1), loc='upper left')
    
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Crear el DataFrame con todos los datos, incluyendo la base "Sin estrategia"
data = {
    'strategy': ['hyde', 'q2d', 'rag_fusion', 'fire_doc', 'fire_query', 'fire_answer', 'doc+answer'] * 3 +
                ['Sin estrategia'] * 3,
    'dataset': ['nfcorpus'] * 7 + ['fiqa'] * 7 + ['scifact'] * 7 +
               ['nfcorpus', 'fiqa', 'scifact'],
    'input': [27.33, 1713.82, 106.33, 2166.86, 2127.86, 2166.86, 4333.72,
              35.67, 1197.96, 114.67, 1294.59, 1261.59, 1294.59, 2589.18,
              47.64, 1922.96, 123.64, 2244.19, 2212.19, 2244.19, 4488.38,
              0, 0, 0],
    'output': [1713.82, 444.05, 169.42, 357.7, 57.41, 177.03, 534.73,
               2112.8, 461.42, 81.91, 381.33, 74.15, 140.91, 522.24,
               2528.56, 304.49, 107.96, 281.76, 61.48, 363.81, 645.57,
               0, 0, 0],
    'nDCG10': [38.1, 30.8, 34.8, 39.9, 32.8, 40.1, 40.6,
               45.9, 43.6, 39.1, 45.9, 37.0, 44.9, 45.9,
               75.5, 78.1, 72.6, 75.9, 68.9, 76.3, 75.6,
               36.1, 44.3, 74.0]
}
df = pd.DataFrame(data)

# 2. Calcular el costo
df['cost_per_1k'] = ((df['input'] * 0.5) + (df['output'] * 1.5)) / 1000

# 3. Definir colores, formas y valores base
strategies = df['strategy'].unique()
datasets = df['dataset'].unique()
color_palette = sns.color_palette('tab10', n_colors=len(strategies))
color_map = dict(zip(strategies, color_palette))
marker_map = {'nfcorpus': 'o', 'fiqa': 's', 'scifact': '^'}

# Diccionario con los valores de nDCG@10 para la línea base
baseline_ndcg = {'nfcorpus': 36.1, 'fiqa': 44.3, 'scifact': 74.0}

# 4. Generar una gráfica para cada dataset
for dataset_name in datasets:
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(10, 7))

    df_subset = df[df['dataset'] == dataset_name]

    # Dibujar la línea base horizontal
    base_value = baseline_ndcg[dataset_name]
    ax.axhline(y=base_value, color='red', linestyle='--', linewidth=2, label=f'Baseline nDCG@10 ({base_value})')

    sns.scatterplot(
        data=df_subset,
        x='cost_per_1k',
        y='nDCG10',
        hue='strategy',
        palette=color_map,
        style='dataset',
        markers=marker_map,
        s=150,
        ax=ax
    )

    # Anotar los puntos
    for i, row in df_subset.iterrows():
        ax.text(row['cost_per_1k'] + 0.04, row['nDCG10'], row['strategy'], fontsize=9)

    ax.set_title(f'Costo vs. nDCG@10 para: {dataset_name.upper()}', fontsize=16, weight='bold')
    ax.set_xlabel('Costo por 1,000 Ejecuciones ($)', fontsize=12)
    ax.set_ylabel('nDCG@10', fontsize=12)
    ax.legend(title='Leyenda', bbox_to_anchor=(1.02, 1), loc='upper left')

    plt.tight_layout()
    plt.show()
```
